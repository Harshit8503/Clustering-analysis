```python
#intializing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("Dataset.csv") #reading csv file
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum() #checking null values 
```

```python
#We can observe that MINIMUM_PAYMENTS has 313 missing values and CREDIT_LIMIT has 1
#To fill these MISSING VALUES
data['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].mean(), inplace=True)
data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].mean(), inplace=True)
#We replace them with mean value of the particular column
```

```python
scaled_data = data.drop(columns=['CUST_ID'],axis=1)  # Dropping 'CUST_ID' as it's not needed for clustering
```

```python
scaled_data
```

```python
#Feature Scaling 
scaler=StandardScaler()
scaled_data=pd.DataFrame(scaler.fit_transform(scaled_data), columns=scaled_data.columns)
```

```python
#Data Visualization
#We check correlated values on the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(scaled_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Distribution of the 'BALANCE' feature
plt.figure(figsize=(8, 6))
sns.histplot(scaled_data['BALANCE'], kde=True, color='blue')
plt.title('Distribution of BALANCE')
plt.show()
```

```python
# Boxplot for 'PURCHASES' to identify outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=scaled_data['PURCHASES'])
plt.title('Boxplot of PURCHASES')
plt.show()
```

```python
#To check pairplot
sns.pairplot(scaled_data[['BALANCE', 'PURCHASES', 'CASH_ADVANCE', 'CREDIT_LIMIT']])
plt.title('Pairplot of Key Features')
plt.show()
```

```python
#To implement clustering
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)  # Set k=4 (can be adjusted based on silhouette score)
kmeans_labels = kmeans.fit_predict(scaled_data)
```

```python
#To Evaluate we use Silhouette score
silhouette_avg = silhouette_score(scaled_data, kmeans_labels)
print(f'Silhouette Score for K-Means: {silhouette_avg}')
```

```python
#To add cluster labels 
data['Cluster'] = kmeans_labels
```

```python
#statistics data from clusters
for cluster_num in range(4):  # Assuming 4 clusters from K-Means
    print(f"\nCluster {cluster_num} Statistics:\n")
    print(data[data['Cluster'] == cluster_num].describe())
```

```python
from sklearn.cluster import DBSCAN

# DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=15)
dbscan_labels = dbscan.fit_predict(scaled_data)
```

```python
# Evaluate DBSCAN using silhouette score
silhouette_dbscan = silhouette_score(scaled_data, dbscan_labels)
print(f'Silhouette Score for DBSCAN: {silhouette_dbscan}')
```

```python
## In this case, Kmeans gives better score than DBSCAN
```

```python
## Lets try out using PCA
```

```python
from sklearn.decomposition import PCA
# apply PCA to reduce to 2 dimensions before clustering
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# K-Means on PCA-reduced data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
kmeans_pca_labels = kmeans_pca.fit_predict(pca_data)

# Visualize clusters in 2D PCA space
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1], hue=kmeans_pca_labels, palette='Set1', s=60)
plt.title('PCA-reduced Clusters(K-means)')
plt.show()

# Evaluate with Silhouette Score after PCA
silhouette_pca = silhouette_score(pca_data, kmeans_pca_labels)
print(f'Silhouette Score after PCA: {silhouette_pca}')
```

```python
pca_dbscan = DBSCAN(eps=0.5, min_samples=15)
pca_dbscan_labels = pca_dbscan.fit_predict(scaled_data)

# Visualize clusters in 2D PCA space
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1], hue=pca_dbscan_labels, palette='Set1', s=60)
plt.title('PCA-reduced Clusters(DBSCAN)')
plt.show()

# Evaluate with Silhouette Score after PCA
silhouette_pca = silhouette_score(pca_data, pca_dbscan_labels)
print(f'Silhouette Score after PCA: {silhouette_pca}')
```

```python
# Without PCA  K-Means Score= 0.2505 DBSCAN Score=-0.1372
#With PCA K-Means Score= 0.4523 DBSCAN Score=-0.2911

##Recommendation : Use K-means for this case as higher the Score better the clustering.
```
